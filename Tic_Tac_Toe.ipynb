```python
def tic_Tac_Toe():
    a1 = 3
    b1 = 3
    board = [[0 for x in range(a1)] for y in range(b1)]
    i = 0
    j = 0
    while(i < len(board[0])):
        j = 0
        while(j < len(board[i])):
            board[j][i] = str(board[j][i])
            print(str(board[j][i]) + " ", end = "")
            j+=1
        print("")
        i+=1
    c = 0
    print()
    while(c < 9):
        if(c % 2 == 0):
            print("X's turn: ")
            isValid = False
            while(isValid != True):
                row = int(input("Enter row (between 0 and 2): "))
                col = int(input("Enter column (between 0 and 2): "))
                if(row > 2):
                    print("Invalid row value. Enter new row and column. ")
                    continue
                if(col > 2):
                    print("Invalid column value. Enter new row and column. ")
                    continue
                if(board[row][col] != "0"):
                    if(board[row][col] != "0"):
                        print("Slot taken. Enter new row and column. ")
                else:
                    isValid = True
            i = 0
            j = 0
            board[row][col] = 'X'
            while(i < len(board[0])):
                j = 0
                while(j < len(board[i])):
                    print(str(board[i][j]) + " ", end = "")
                    j+=1
                print("")
                i+=1
            c+=1
        else:
            print("O's turn: ")
            isValid = False
            while(isValid != True):
                row = int(input("Enter row (between 0 and 2): "))
                col = int(input("Enter column (between 0 and 2): "))
                if(row > 2):
                    print("Invalid row value. Enter new row and column. ")
                    continue
                if(col > 2):
                    print("Invalid column value. Enter new row and column. ")
                    continue
                if(board[row][col] != "0"):
                    if(board[row][col] != "0"):
                        print("Slot taken. Enter new row and column. ")
                else:
                    isValid = True
            i = 0
            j = 0
            board[row][col] = 'O'
            while(i < len(board[0])):
                j = 0
                while(j < len(board[i])):
                    print(str(board[i][j]) + " ", end = "")
                    j+=1
                print("")
                i+=1
            c+=1
        if(is_X_Winner(board)):
            print("X Wins!")
            c = 10
        if(is_O_Winner(board)):
            print("O Wins!")
            c = 10
    if(c == 9):
        print("Draw.")
```

```python
def is_X_Winner(board):
    coords = [[0,0,0],[1,1,1],[2,2,2],[0,1,2],[2,1,0]]
    col1 = board[0][0] + board[1][0] + board[2][0]
    col2 = board[0][1] + board[1][1] + board[2][1]
    col3 = board[0][2] + board[1][2] + board[2][2]
    if(col1 == "XXX"):
        return True
    if(col2 == "XXX"):
        return True
    if(col3 == "XXX"):
        return True
    i = 0
    while(i < len(coords)):
        j = 0
        a = ""
        while(j < len(coords[i])):
            if(board[coords[i][j]][j] != "0"):
                a = a + board[coords[i][j]][j]
            if(a == "XXX"):
                return True
            j+=1
#         print(a)
        i+=1
    return False
```

```python
def is_O_Winner(board):
    coords = [[0,0,0],[1,1,1],[2,2,2],[0,1,2],[2,1,0]]
    col1 = board[0][0] + board[1][0] + board[2][0]
    col2 = board[0][1] + board[1][1] + board[2][1]
    col3 = board[0][2] + board[1][2] + board[2][2]
    if(col1 == "OOO"):
        return True
    if(col2 == "OOO"):
        return True
    if(col3 == "OOO"):
        return True
    i = 0
    while(i < len(coords)):
        j = 0
        a = ""
        while(j < len(coords[i])):
            if(board[coords[i][j]][j] != "0"):
                a = a + board[coords[i][j]][j]
            if(a == "OOO"):
                return True
            j+=1
#         print(a)
        i+=1
    return False
```
